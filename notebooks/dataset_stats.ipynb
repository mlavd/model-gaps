Imports

```python
from pathlib import Path
from rich.progress import track
import pandas as pd
```

Load dataset stats.

```python
stats = []

for dataset in track(list(Path('../data/jsonl').iterdir())):
    # if 'task' in dataset.stem: continue
    
    train = pd.read_json(dataset.joinpath('train.jsonl'), lines=True)
    test = pd.read_json(dataset.joinpath('test.jsonl'), lines=True)
    valid = pd.read_json(dataset.joinpath('valid.jsonl'), lines=True)
    
    stats.append({
        'dataset': dataset.stem,
        'train_pos': train.target.sum(),
        'train_neg': (1 - train.target).sum(),
        'valid_pos': valid.target.sum(),
        'valid_neg': (1 - valid.target).sum(),
        'test_pos': test.target.sum(),
        'test_neg': (1 - test.target).sum(),
    })

stats = pd.DataFrame(stats)
stats
```

```python
stats['total_pos'] = stats.train_pos + stats.valid_pos + stats.test_pos
stats['total_neg'] = stats.train_neg + stats.valid_neg + stats.test_neg
stats
```
